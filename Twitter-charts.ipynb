```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
df_doge = pd.read_csv("tweets_doge_all2.csv", low_memory = False)
```

```python
df_doge["date"].count()
```

```python
df_doge = df_doge[["date", "username"]]
```

```python
df_doge["date"] = pd.to_datetime(df_doge["date"])

df_doge = df_doge.sort_values(by="date", ascending=False)
```

```python
x_axis = df_doge["date"].unique()
y_axis = df_doge["date"].value_counts()

fig, ax = plt.subplots(figsize=(15,10))
ax.plot(x_axis, y_axis);
ax.set(title = "Daily 'Doge' Tweets 1/1/21 to 4/30/21",
       xlabel = "Date",
       ylabel = "Daily Tweets");
```
