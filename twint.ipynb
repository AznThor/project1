```python
# Import Twint scraper and Pandas
import twint
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Configure Twint
c = twint.Config()
c.Username = "elonmusk"
c.Search = "doge" or "dogecoin" or "#doge" or "#dogecoin"
c.Since = "2020-1-1"
c.Until = "2021-4-30"
c.Pandas = True
c.Hide_output = True

# Run
twint.run.Search(c)

# Store data in DataFrame
Tweets_df = twint.storage.panda.Tweets_df
```

```python
# Display all Elon Musk tweets about Doge by dates and number of likes
df_tweets_doge_elon = Tweets_df[["date", "username", "nlikes"]]
df_tweets_doge_elon
```
